convert this [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]  to this form ['Action','Adventure','Fantasy','Science Fiction'] using a helper function convert

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'][0]
```

only top 3 actors are required

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
movies['cast']
```

only director of the movie  required

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])

    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

applying a transformation for without space

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['overview']=movies['overview'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

creating a new column tags by concatenate overview, genres, keywords, cast,crew

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])#removing overview, genres, keywords, cast,crew columns
new.head().shape
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x)) #converting into string
new.head()
```

```python
new['tags'][0]
```

```python
#converting text to vector (text vectorization)
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags']).toarray()
vector #each movie is a vector
```

```python
vector.shape
```

```python
vector[0]
```

distance inversely proportional to similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
similarity
```

```python
similarity[1]# it shows the similarity of 2nd movie with the other movies
```

```python
def recommend(movie):
  movie_index =new[new['title']==movie].index[0]
  distance= similarity[movie_index]
  movies_list =sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
